```python
file = "/home/rpinto/Downloads/WS21/KlimaDynamics/Exercises/Klimadynamics/data/nino34.long-1870-2020.data"
nina34 = np.loadtxt(file, skiprows=1, max_rows=2020-1870+1)  # [year, nino34Jan, ..., nino34Dec]
nina34[nina34 < -90] = np.nan
nina34[:, 1:] = nina34[:, 1:] - np.nanmean(nina34[:, 1:])  # remove long-term mean

# to xarray
ds_nina34 = xr.Dataset()
ds_nina34.coords['time'] = pd.date_range('1870-01-01', '2020-12-31', freq='1m').values
ds_nina34['nina34'] = (('time', ), nina34[:, 1:].ravel())
```

```python
# calculate annual nina34 index
ds_nina34_ann = ds_nina34.groupby('time.year').mean('time')

nina34_std = ds_nina34_ann.nina34.std()

ds_nina34_ann['el_nino'] = (('year', ), ds_nina34_ann.nina34 > nina34_std)  # el nino years
ds_nina34_ann['la_nina'] = (('year', ), ds_nina34_ann.nina34 < -nina34_std)  # la nina years

year_la_nina = ds_nina34_ann.year[ds_nina34_ann.la_nina]
year_el_nino = ds_nina34_ann.year[ds_nina34_ann.el_nino]
```

```python
#%% plot nina34 time series
fig = plt.figure(figsize=(7, 4), constrained_layout=True)

plt.plot(ds_nina34_ann.year, ds_nina34_ann.nina34, color='k', marker='x', linewidth=2, label='NINA3.4')
plt.scatter(year_la_nina, ds_nina34_ann.nina34.sel(year=year_la_nina), color='blue', linewidth=4, label='La Nina')
plt.scatter(year_el_nino, ds_nina34_ann.nina34.sel(year=year_el_nino), color='red', linewidth=4, label='El Nino')

plt.axhline(y=0, color='k', linestyle=':')
plt.axhline(y=nina34_std, color='k', linestyle='--', label='$\pm \sigma$')
plt.axhline(y=-nina34_std, color='k', linestyle='--')

plt.xlabel('Year')
plt.ylabel('NINA3.4')

plt.legend(ncol=4)
```

```python
slp_djf = ds.sel(time=ds['time.season']=='DJF').groupby('time.year').mean('time') #average over time
slp_reg = slp_djf.isel(lat=(slp_djf.lat <= 20) & (slp_djf.lat >= -20)) #select region
```

```python
n_lat = len(slp_reg.lat)
n_lon = len(slp_reg.lon)
m = len(slp_reg.year)
dim_new = (m, n_lat*n_lon)
```

```python
slp_reg_stack = slp_reg.stack(latlon=('lat', 'lon'))

# weights
W = np.diag(np.sqrt(np.cos(np.deg2rad(slp_reg_stack.latlon.lat))))

# calculate covariance matrix
# calculate anomalies D and weigh them
D = slp_reg_stack - slp_reg_stack.mean('year')
D_weighted = D.slp.values @ W

cov_mat = (1/m) * D.slp.T.values @ D.slp.values
cov_mat_weighted = (1/m) * D_weighted.T @ D_weighted
```

```python
# PCA
# calculate eigenvec and eigenval
# column eig_val[:,i] is the eigenvector corresponding to the eigenvalue eig_vec[i]
l, E = eig(cov_mat_weighted)  # eigen_val, eigen_vec
l = l.real
E = E.real

ix = np.argsort(l)
l = l[ix][::-1]
E = E[:, ix][:, ::-1]
```

```python
# plot E and l
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(6, 5), sharex=True)

# plot eigenvalues
ax1.plot(range(1, 1+len(l)), l, 'ok-')
ax1.set_ylabel('Eigenvalue')
ax1.set_xlabel("Number of Eigen vectors")

# plot explained variance
ex_var = np.cumsum(l)/np.sum(l)
ax2.plot(range(1, 1+len(l)), ex_var, 'or-')
ax2.set_ylim([0.85, 1.01])
ax2.set_ylabel('Explained variance')
ax2.set_xlim([0, 10])
ax2.set_xlabel("Number of Eigen vectors")
fig.tight_layout()
```

II. 

In general, the number of grid point in your region will largely exceed the number of El Niño and La Niña years. So an a-priori reduction of spatial DOF is required. Use the first $\tilde{q}$ EOFs for the reduction, i.e. work on the respective PCs instead. Perform the hierarchical Hotelling T$^2$-test, i.e. let $\tilde{q}$ go from 1 to 10 
(note that if q̃ = 1 the covariance matrix reduced to a variance) and assess the hypotheses for each $\tilde{q}$

  $H_{0}$: $E(\vec{a}^{(c)})$

  $H_{A}$: $E(\vec{a}^{(\omega)})\neq E(\vec{a}^{(c)})$
    
where $\omega$ indicates warm (La Niña) and $\textit{c}$ cold (El Niño) conditions.
 
   i Define error level $\alpha$ and derive critical values for T$^2$ for $\tilde{q}$ = 1, . . . 10.
    
   ii Estimate T$^2$ test variable from samples for $\tilde{q}$ = 1, . . . 10.
    
   iii Plot values of [i] and [ii] as function of $\tilde{q}$

```python
#%% test difference for significance
# with PC
# with expansion coefficients
# composit for expansion coefficients
# difference of expansion coeff
# significance analysis of expansion coeff 
# T square
# test statistic

q_tilde_arr = np.arange(1, 20)
alpha_arr = np.array([0.001, 0.01, 0.05, 0.1])

T_q_squared_mat = np.zeros(shape=len(q_tilde_arr))
T_q_crit_squared_mat = np.zeros(shape=(len(q_tilde_arr), len(alpha_arr)))

# anomalies (D)
mx = len(year_la_nina)
my = len(year_el_nino)
Dx = D.sel(year=year_la_nina).mean('year')
Dy = D.sel(year=year_el_nino).mean('year') 
```

```python
# calculate T squared values
for i, q_tilde in enumerate(q_tilde_arr):

    G = E[:, :q_tilde]  # guess vectors (first q eigenvectors)
    S_tilde_inv = np.diag(1/l[:q_tilde])  # S matrix with eigenvalues on diag

    print(S_tilde_inv.shape)

    # principle components
    ax = G.T @ Dx.slp.values
    ay = G.T @ Dy.slp.values

    # estimate T square from samples q=1...10
    T_q_squared_mat[i] = mx*my/(mx+my)*(ax - ay).T @ S_tilde_inv @ (ax - ay)
```

```python
# calculate critical values
for i, q_tilde in enumerate(q_tilde_arr):
    for j, alpha in enumerate(alpha_arr):
        # critical value from alpha 
        Z_crit = f.ppf(q=1-alpha, dfn=q_tilde, dfd=mx + my - q_tilde - 1)
        T_q_crit_squared_mat[i, j] = q_tilde*(mx+my-2)/(mx+my-q_tilde-1) * Z_crit
```

```python
fig = plt.figure()

lst = ['-.', ':', '--', '-']
for i, alpha in enumerate(alpha_arr):
    plt.plot(q_tilde_arr, T_q_crit_squared_mat[:, i], 'k'+lst[i], label=alpha)
    
plt.plot(q_tilde_arr, T_q_squared_mat, 'or-')
    
plt.legend()
```
