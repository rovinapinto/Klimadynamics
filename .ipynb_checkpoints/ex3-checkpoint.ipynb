```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import netCDF4 as nc
import xarray as xr
from scipy.linalg import eig
```

Read HadSLP2 (Allan & Ansell, 2006) gridded global sea level pressure (SLP) anomalies from 1850 to 2004. Again extract the December months (alternatively you may average over the December to February (DJF) season). Plot the longterm mean SLP field for December (or DJF).

I. Calculate zonal mean SLP anomalies, which is a 2-dimensional array with $\mathrm{nlat}$ number of latitudes and $\mathrm{nt}$ number of times (December or DJF). You may denote the anomaly matrix (Matrix der zentrierten Differenzen) according to the lecture as $\mathrm{D}$.

- Plot $\mathrm{D}$ in a so-called Hovmoeller diagram, with latitudes on x-axis and time on y-axis. What do you see?

- Plot the respective covariate matrix estimated as $\frac{1}{m}D D^T$. What do you see?

```python
def get_multivariate_normal(x, mu, cov_mat):
    """
    Gaussian multivariate PDF
    
    x:          array with elements of random variable (dim: q)
    mu:         array with mean of random variable (dim: q)
    cov_mat:    covariance matrix (dim: q x q)
    """
    
    q = cov_mat.shape[0]
    
    Z = np.sqrt((2*np.pi)**q*det(cov_mat))
    
    prob = 1/Z * np.exp(-0.5*(x-mu).T @ inv(cov_mat) @ (x-mu))
    
    return prob
```

```python
def est_mean(x):
    
    m = len(x)
    mean = 1/m * np.sum(x)
    
    return mean    
```

```python
def est_cov_mat(x1, x2):
    """
    np.cov(slp_lis.values, slp_rey.values)
    """

    m = len(x1)
    
    assert m == len(x2)
    
    mean1 = 1/m * np.sum(x1)
    mean2 = 1/m * np.sum(x2)
    
    d_1 = x1 - mean1
    d_2 = x2 - mean2
    
    D = np.array([d_1, d_2])
    
    cov_mat = 1/m * np.inner(D, D)
    
    return cov_mat
```

```python
with xr.open_dataset("HadSLP2_slp-mnmean-real.nc") as ds:
    print(ds.keys())
```

```python
slp_djf = ds.sel(time=ds['time.season']=='DJF').groupby('time.year').mean('time')
```

```python
m = len(slp_djf.year)
    
# average over longitude
slp_avg_lon = slp_djf.mean('lon')

# avergae over longitude and every year
slp_avg_lon_time = slp_avg_lon.mean('year')

# calculate zonal SLP anomalies
D = slp_avg_lon - slp_avg_lon_time
```

```python
# plot D /Hovmoeller diagram)
norm = mcolors.TwoSlopeNorm(vmin=np.min(D.slp.values),vcenter=0, vmax=np.max(D.slp.values))

fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(4, 9), gridspec_kw={'height_ratios': [1, 0.1]})
im = ax1.pcolormesh(D.lat.values, D.year.values, D.slp.values, cmap='RdBu_r', shading='nearest', norm=norm)
ax2.plot(slp_avg_lon_time.lat.values, slp_avg_lon_time.slp.values)
fig.colorbar(im, ax=ax1, orientation='horizontal')
ax1.set_xlabel('lat')
ax1.set_ylabel('year')
ax2.set_xlabel('lat')
ax2.set_ylabel('sea level pressure')
fig.tight_layout()
```

```python
# covariance matrix
# weight covariance matrix with area
W = np.diag((np.cos(D.lat*np.pi/180)))

D_weighted = D.slp.values @ W

cov_mat = (1/m) * D.slp.T.values @ D.slp.values
cov_mat_weighted = (1/m) * D_weighted.T @ D_weighted
```

```python
norm = mcolors.TwoSlopeNorm(vmin=np.min(cov_mat), vcenter=0, vmax=np.max(cov_mat))
fig, axes = plt.subplots(1, 2)
axes[1].set_title('Covariance matrix based\non weighted $D$')
axes[0].set_title('Covariance matrix based\non original $D$')
im0 = axes[0].imshow(cov_mat, cmap='seismic', norm=norm)
im1 = axes[1].imshow(cov_mat_weighted, cmap='seismic', norm=norm)
fig.colorbar(im0, ax=axes, orientation='horizontal') 
```

II.
Calculate and plot the results of a principal component analysis.

- Plot the eigenvalues as well as the explained variances of the covariance matrix as estimated in I. (explained variance: the fraction of variance explained by the first k eigenvalues as a function of k).
- Plot the first two eigenvectors. What do they tell us?
- Plot the corresponding principal components of the first two eigenvectors

```python
# PCA
# calculate eigenvec and eigenval
# column eig_val[:,i] is the eigenvector corresponding to the eigenvalue eig_vec[i]
l_0, E_0 = eig(cov_mat_weighted)  # eigen_val, eigen_vec
ix = np.argsort(l_0)
l = l_0[ix][::-1]
E = E_0[:, ix][:, ::-1] 

#unweighted
l_0, E_0 = eig(cov_mat)  # eigen_val, eigen_vec
ix = np.argsort(l_0)
l_nw = l_0[ix][::-1]
E_nw = E_0[:, ix][:, ::-1] 
```

```python
fig, [ax1, ax2, ax3, ax4] = plt.subplots(1, 4, figsize=(14, 5), sharex=True)

# plot eigenvalues
ax1.plot(range(1, 1+len(l)), l, 'ok-')
ax1.set_ylabel('Eigenvalue')
ax1.set_xlabel('Number of eigenvalue')

# plot explained variance
ex_var = np.cumsum(l)/len(l)
ax2.plot(range(1, 1+len(l)), ex_var, 'or-')
ax2.set_ylim(bottom=0)
ax2.set_ylabel('Explained variance')
ax2.set_xlabel('Number of eigenvalue')

# plot unweighted eigenvalues
ax3.plot(range(1, 1+len(l_nw)), l_nw, 'ok-')
ax3.set_ylabel('Unweighted Eigenvalue')
ax3.set_xlabel('Number of eigenvalue')

# plot unweighted explained variance
ex_var = np.cumsum(l_nw)/len(l_nw)
ax4.plot(range(1, 1+len(l_nw)), ex_var, 'or-')
ax4.set_ylim(bottom=0)
ax4.set_ylabel('Unweighted Explained variance')
ax4.set_xlabel('Number of eigenvalue')

fig.tight_layout()
```

```python
# plot first two eigenvalues
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# first and second eigenvector
axes[0,0].plot(D.lat.values, E[:, 0], label='First eigenvector')
axes[0,0].plot(D.lat.values, E[:, 1], label='Second eigenvector')
axes[0,0].legend()
axes[0,0].set_xlabel('Latitude')
axes[0,0].set_ylabel('Component of eigenvector')

# principle components of first and second eigenvector
# calculate principle components (A) using the time series as X
A = E.T @ D_weighted.T  # principal components

for i in range(2):
    print(np.var(A[i, :]), l[i])

axes[0,1].plot(D.year.values, A[0, :], label='PCA of first eigenvector')
axes[0,1].plot(D.year.values, A[1, :], label='PCA of second eigenvector')
axes[0,1].legend()
axes[0,1].set_xlabel('Year')
axes[0,1].set_ylabel('Principal component')

#unweighted
# first and second eigenvector
axes[1,0].plot(D.lat.values, E_nw[:, 0], label='First eigenvector')
axes[1,0].plot(D.lat.values, E_nw[:, 1], label='Second eigenvector')
axes[1,0].legend()
axes[1,0].set_xlabel('Latitude')
axes[1,0].set_ylabel('Component of unweighted eigenvector')

# principle components of first and second eigenvector
# calculate principle components (A) using the time series as X
A_nw = E_nw.T @ D.slp.T.values  # principal components

for i in range(2):
    print(np.var(A_nw[i, :]), l[i])

axes[1,1].plot(D.year.values, A_nw[0, :], label='PCA of first eigenvector')
axes[1,1].plot(D.year.values, A_nw[1, :], label='PCA of second eigenvector')
axes[1,1].legend()
axes[1,1].set_xlabel('Year')
axes[1,1].set_ylabel('Principal component')

fig.tight_layout()
```

III. Calculate and plot the filtered variability in terms of filtered $ \mathrm{D}_{fil} = \mathrm{\hat{U}}\mathrm{\hat{A}} = \mathrm{\hat{U}}\mathrm{\hat{U}^\mathrm{T}}\mathrm{D}$

- Plot the filter $\mathrm{\hat{U}}\mathrm{\hat{U}^\mathrm{T}}$
- Plot the Hovmoeller diagram of the filtered variability and the differences

IV. Repeat analysis with weights such that the variance (covariance) is weighted with the area that is represented by the grid points, which is proportional to the cosine of the geographic latitude. This is performed by multiplication of $\mathrm{D}$ with a weight matrix $ diag (\sqrt{\cos(\varphi_{j})})$

```python
# filetered components: only first two eigenvalues
# not weighted
U_tilde_nw = E_nw[:, :2]
filtr_nw = U_tilde_nw @ U_tilde_nw.T
D_fil_nw = (filtr_nw @ D.slp.transpose('lat', 'year').values).T

# weighted
U_tilde = E[:, :2]
filtr = U_tilde @ U_tilde.T
D_fil = (filtr @ D_weighted.T).T
```

```python
# plot filter
norm = mcolors.TwoSlopeNorm(vmin=np.min(filtr), vcenter=0, vmax=np.max(filtr))
fig, axes = plt.subplots(1, 2)
axes[0].set_title('Filter $U U^T$ based\non weighted $D$')
axes[1].set_title('Filter $U U^T$ based\non original $D$')
im0 = axes[0].imshow(filtr, cmap='RdBu_r', norm=norm)
im1 = axes[1].imshow(filtr_nw, cmap='RdBu_r', norm=norm)
fig.colorbar(im0, ax=axes, orientation='horizontal')

# Hovmoeller plot: original D, weighted D, filtered D, filter - weighted D
fig, axes = plt.subplots(1, 6, figsize=(12, 5), sharex=True, sharey=True, constrained_layout=True)

fig.suptitle('Hovmoeller')
titles = ['Pressure anomaly\n$D$', 'Weighted\npressure anomaly\n$D_w$', 'Filtered\npressure anomaly\n$D_f$',
          'Filtered and weighted\npressure anomaly\n$D_{fw}$', 
          'Filter effect\n$D_w - D_{fw}$', 'Weight effect\n'+r'$D_w - D$']

for i, ax in enumerate(axes):
    ax.set_title(titles[i])
    ax.set_xlabel('lat')
axes[0].set_ylabel('year')

norm = mcolors.TwoSlopeNorm(vmin=np.min(D_fil), vcenter=0, vmax=np.max(D_fil))
kwargs = dict(shading='nearest', norm=norm, cmap='RdBu_r')

# plot Hovmollers
im = axes[0].pcolormesh(D.lat.values, D.year.values, D.slp.values, **kwargs)
im = axes[1].pcolormesh(D.lat.values, D.year.values, D_weighted, **kwargs)
im = axes[2].pcolormesh(D.lat.values, D.year.values, D_fil_nw, **kwargs)
im = axes[3].pcolormesh(D.lat.values, D.year.values, D_fil, **kwargs)
im = axes[4].pcolormesh(D.lat.values, D.year.values, D_weighted-D_fil, **kwargs)
im = axes[5].pcolormesh(D.lat.values, D.year.values, D_weighted-D.slp.values, **kwargs)

fig.colorbar(im, ax=axes, shrink=0.6, label='Pressure Anomaly [hPa]')
```
