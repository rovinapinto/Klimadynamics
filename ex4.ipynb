```python
# get dimensions
n_year = 170
n_lat = 9
n_lon = 72
dim_new = (n_year, n_lat*n_lon)
```

```python
slp_reg_stack = slp_reg.stack(latlon=('lat', 'lon'))
# slp_reg_stack.unstack('latlon')  # retransformation
```

```python
# weights
W = np.diag(np.sqrt(np.cos(np.deg2rad(slp_reg_stack.latlon.lat))))
m = len(slp_djf.year)

# calculate covariance matrix
# calculate anomalies D and weigh them
D = slp_reg_stack - slp_reg_stack.mean('year')
D_weighted = D.slp.values @ W

cov_mat = (1/m) * D.slp.T.values @ D.slp.values
cov_mat_weighted = (1/m) * D_weighted.T @ D_weighted
```

```python
# PCA
# calculate eigenvec and eigenval
# column eig_val[:,i] is the eigenvector corresponding to the eigenvalue eig_vec[i]
l_0, E_0 = eig(cov_mat_weighted)  # eigen_val, eigen_vec
l_0 = l_0.real
E_0 = E_0.real
ix = np.argsort(l_0)
l = l_0[ix][::-1]
E = E_0[:, ix][:, ::-1] 

#unweighted
l_0, E_0 = eig(cov_mat)  # eigen_val, eigen_vec
l_0 = l_0.real
E_0 = E_0.real
ix = np.argsort(l_0)
l_nw = l_0[ix][::-1]
E_nw = E_0[:, ix][:, ::-1] 

# calculate principle components of two largest eigenvalues
n = 2
D_fil = (E[:, :n] @ E[:, :n].T @ D_weighted.T).T
```

```python
# plot E and l
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(6, 5), sharex=True)

# plot eigenvalues
ax1.plot(range(1, 1+len(l_nw)), l_nw, 'ok-')
ax1.set_ylabel('Eigenvalue')
ax1.set_xlabel("Number of Eigen vectors")

# plot explained variance
ex_var = np.cumsum(l_nw)/np.sum(l)
ax2.plot(range(1, 1+len(l_nw)), ex_var, 'or-')
ax2.set_ylabel('Explained variance')
ax2.set_xlim([0, 10])
ax2.set_xlabel("Number of Eigen vectors")
fig.tight_layout()
```

```python
# plot first two eigenvalues
fig, [ax1,ax2] = plt.subplots(2, 1, figsize=(10,6))

# first and second eigenvector
ax1.plot(D.lat.values, E[:, 0], label='First eigenvector')
ax1.plot(D.lat.values, E[:, 1], label='Second eigenvector')
ax1.legend()
ax1.set_xlabel('Latitude')
ax1.set_ylabel('Component of eigenvector')

# principle components of first and second eigenvector
# calculate principle components (A) using the time series as X
A = E.T @ D_weighted.T  # principal components

for i in range(2):
    print(np.var(A[i, :]), l[i])

ax2.plot(D.year.values, A[0, :], label='PCA of first eigenvector')
ax2.plot(D.year.values, A[1, :], label='PCA of second eigenvector')
ax2.legend()
ax2.set_xlabel('Year')
ax2.set_ylabel('Principal component')

fig.tight_layout()
```
